```python
import numpy as np
import matplotlib.pyplot as plt

h = 10.0/101
t = np.linspace(0,10,101)
v = [0] * 101
ay = [0] * 101

for i in range(1,101,1):
  t[i] = h*(i) + t[0]

for i in range(1,101):
  ay[i] = t[i]/160 * -((t[i])**3 + 1568)

for j in range(len(t)-1):
  n = 100
  x = [0]* n
  x[0] = 1
  c = 0
  for i in range(n-1):
    c = ((3*h*(x[i]**2))/40) + 1
    x[i+1] = x[i] - (x[i] + (9.8*h)-(v[j])+(h/40*x[i]**3))/c
  v[j+1] = x[n-1]


print(v)
print(ay)
plt.plot(t,v)
plt.plot(t,ay)
```
